```python
def sgn(x):
    return 2*(x>=0) - 1
```

```python
def add_noise(y , p=0.2):
    x = copy.deepcopy(y)
    mask = np.random.choice(len(x), size=int(p*len(x)), replace=False)
    x[mask] = -1 * x[mask]
    return x

def add_missing(y, p=0.2):
    x = copy.deepcopy(y)
    mask = np.random.choice(len(x), size=int(p*len(x)), replace=False)
    x[mask] = 0
    return x

```

```python
p = 0.2
s = 0
n = 10000
for _ in range(n):
    A_noisy = add_noise(A , p)
    B_noisy = add_noise(B , p)
    C_noisy = add_noise(C , p)

    apred = sgn(np.matmul(A_noisy.T, W)).T
    bpred = sgn(np.matmul(B_noisy.T, W)).T
    cpred = sgn(np.matmul(C_noisy.T, W)).T

    if sum(apred == a) == len(a) and sum(bpred == b) == len(b) and sum(cpred == c) == len(c):
       s += 1
acc = s/n
acc
```

```python
ps = np.arange(0, 1 ,0.01)
n = 1000
acc1 = []
acc2 = []

for p in ps:
    s1 = 0
    s2 = 0
    for _ in range(n):
        A_noisy = add_noise(A , p)
        B_noisy = add_noise(B , p)
        C_noisy = add_noise(C , p)

        apred = sgn(np.matmul(A_noisy.T, W)).T
        bpred = sgn(np.matmul(B_noisy.T, W)).T
        cpred = sgn(np.matmul(C_noisy.T, W)).T

        if sum(apred == a) == len(a) and sum(bpred == b) == len(b) and sum(cpred == c) == len(c):
            s1 += 1

        A_noisy = add_missing(A , p)
        B_noisy = add_missing(B , p)
        C_noisy = add_missing(C , p)

        apred = sgn(np.matmul(A_noisy.T, W)).T
        bpred = sgn(np.matmul(B_noisy.T, W)).T
        cpred = sgn(np.matmul(C_noisy.T, W)).T

        if sum(apred == a) == len(a) and sum(bpred == b) == len(b) and sum(cpred == c) == len(c):
            s2 += 1
    acc1 += [s1/n]
    acc2 += [s2/n]
```

```python
plt.plot(ps, acc1, label='noise')
plt.plot(ps, acc2, label='missing')
plt.title('robustness')
plt.xlabel('change percentage')
plt.ylabel('accuracy')
plt.grid()
plt.legend()
```
