```python
def sgn(x):
    return 2*(x>=0) - 1
```

```python
def psgn(x):
    y = 2*(x>=0) - 1
    y[x == 0] = x[x==0]
    return y
```

```python
def ppsgn(x , z):
    y = 2*(x>=0) - 1
    y[x == 0] = z[x==0]
    return y
```

# Q2-3

```python
def add_noise(y , p=0.3):
    x = copy.deepcopy(y)
    mask = np.random.choice(len(x), size=int(p*len(x)), replace=False)
    x[mask] = -1 * x[mask]
    return x

def add_missing(y, p=0.3):
    x = copy.deepcopy(y)
    mask = np.random.choice(len(x), size=int(p*len(x)), replace=False)
    x[mask] = 0
    return x
```

```python
p = add_noise(add_missing(a))
P = add_noise(add_missing(A))
plt.imshow(P.reshape(-1, 16))
plt.title('plane')
plt.show()
plt.imshow(p.reshape(8, -1))
plt.title('plane')
```

# reconstruct

```python
p = ppsgn(np.matmul(P.T, W).T, p)
P = ppsgn(np.matmul(p.T, W.T).T, P)
plt.imshow(P.reshape(-1, 16))
plt.title('plane')
plt.show()
plt.imshow(p.reshape(8, -1))
plt.title('plane')
```

```python
p = add_noise(add_missing(b))
P = add_noise(add_missing(B))
plt.imshow(P.reshape(-1, 16))
plt.title('tank')
plt.show()
plt.imshow(p.reshape(8, -1))
plt.title('tank')
```

```python
p = ppsgn(np.matmul(P.T, W).T, p)
P = ppsgn(np.matmul(p.T, W.T).T, P)
plt.imshow(P.reshape(-1, 16))
plt.title('tank')
plt.show()
plt.imshow(p.reshape(8, -1))
plt.title('tank')
```
